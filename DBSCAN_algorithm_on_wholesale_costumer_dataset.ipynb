```python
#Import the libraries
from sklearn.cluster import DBSCAN #importing dbscan algorithm
from sklearn.preprocessing import StandardScaler #to standardize the data values into a standard format
import numpy as np #importing numpy for scientific calculation 
import pandas as pd #importing pandas library for data handling and manipulation
#importing pyplot module from matplotlib for data visualization including seaborn as sns 
from matplotlib import pyplot as plt 
```

```python
from google.colab import drive #import drive for csv file 
drive.mount('/content/drive', force_remount=True) 
df = pd.read_csv("/content/Wholesale customers data.csv") # reading path of csv file

print(df)
```

```python
print(df.info()) #understanding the data
```

```python
df.drop(['Channel','Region'],axis=1,inplace=True) #dropping the column which are not important(ask domain expert for that)
```

```python
array=df.values #it will remove column name and just returning array
```

```python
array
```

```python
stscaler = StandardScaler().fit_transform(array) #standardizing the values through fit and tranform
```

```python
X =stscaler #storing values in X
```

```python
dbscan = DBSCAN(eps=3, min_samples=7) #dbscan parameters (user_define)
dbscan.fit(X) #fit the data
```

```python
#Noisy samples are given the label -1. 
#as mention in performace 1 is best -1 is wrost 0 is overlapping 
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster']) #these labels will create an column name as cluster and it will store in cl
```

```python
cl #printing the cl 
```

```python
clustered = pd.concat([df,cl],axis=1) #concatinating the two dataframe and storing in new dataframe 
```

```python
clustered[clustered['cluster']==-1] #checking valuess who got the -1 for bad clusters 
```
